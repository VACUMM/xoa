# Compare CROCO and HYCOM outputs

This notebooks uses xoa to facilitate the comparison between CROCO and HYCOM models outputs.  
This involves :
* the automatic discovery of names in dataset thanks the [``xoa.cf``](https://vacumm.github.io/xoa/api/xoa.cf.html) module,
* the conversion from sigma to depths coordinates thanks the ``xoa.sigma`` module,
* the conversion from layer thickness to depths coordinates thanks to the ``xoa.grid`` module,
* the horizontal and vertical regridding to colocate outputs thanks to the ``xoa.regrid`` module.
